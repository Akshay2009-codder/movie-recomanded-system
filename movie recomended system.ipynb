```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credit = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credit.head(1)
```

```python
movies = movies.merge(credit, on='title')
```

```python
movies.head(1)
```

```python
# geners, homepage, keywords, overview, cast, crew
movies.info()
```

```python
movies = movies[["movie_id","genres","keywords","overview","cast","crew","title"]]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def converter (obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i["name"])

    return l
```

```python
movies["genres"] = movies["genres"].apply(converter)
```

```python
movies["keywords"] = movies["keywords"].apply(converter)
```

```python
def converter3 (obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
     if counter != 3 :
         l.append(i["name"])
         counter += 1
     else:
         break
    return l
```

```python
movies["cast"] = movies["cast"].apply(converter3)
```

```python
def fetch_director(obj):
 l = []
 for i in ast.literal_eval(obj):
     if i["job"] == "Director":
         l.append(i["name"])
         break
 return l
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["overview"] = movies["overview"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["crew"] + movies["genres"] + movies["keywords"]
```

```python
new_df = movies[["movie_id","title","tags"]]
```

```python
new_df.head(1)
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```
